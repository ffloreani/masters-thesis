```python
import subprocess

lengths = []
paf = "../data/NCTC86.paf"

def file_line_count(input_file):
    out = subprocess.Popen(
        ['wc', '-l', input_file],
        stdout=subprocess.PIPE,
        stderr=subprocess.STDOUT
    ).communicate()[0]

    return int(out.decode('utf-8').strip().split(' ')[0])


with open(file=paf, mode="r", buffering=1) as paf_handle:

    line_count = file_line_count(paf)

    current_id = "-1"
    current_query_len = 0
    for idx, line in enumerate(paf_handle):
        if idx % 1000 == 0:
            print("\nParsing line {}. Percent complete = {:.2f}\n".format(idx, idx / line_count))

        fields = line.split(sep="\t")

        if current_id == "-1":
            # print("Init first value")
            current_id = fields[0]
            current_query_len = int(fields[1])
        elif current_id == fields[0]:
            # print("Skip entry for the same sequence")
            continue
        else:
            # print("Write down sequence length & set new current sequence")
            lengths.append(current_query_len)
            current_id = fields[0]
            current_query_len = int(fields[1])
```

```python
import matplotlib.pyplot as plt
import numpy as np


average = sum(lengths) // len(lengths)
max_val = max(lengths)
min_val = min(lengths)
print("Max \\ Min \\ Average")
print("{} \\ {} \\ {}".format(max_val, min_val, average))

labels = ["0-1000", "1000-3000", "3000-5000", "5000-10000", "10000-15000", "15000-20000",
                  "20000-+oo"]
y_pos = np.arange(len(labels))

bins = [(0, 1000), (1000, 3000), (3000, 5000), (5000, 10000), (10000, 15000), (15000, 20000)]

bin_count = [0, 0, 0, 0, 0, 0, 0]

assigned = False
for i in lengths:
    for idx, bin in enumerate(bins):
        if bin[0] <= i < bin[1]:
            bin_count[idx] += 1
            assigned = True
            continue

    if not assigned:
        bin_count[6] += 1
    assigned = False

print(y_pos)
print(bin_count)

plt.rcParams.update({'font.size': 15})
plt.rcParams['xtick.major.pad']='15'
plt.rcParams['ytick.major.pad']='15'

plt.figure(figsize=(15,10))
plt.bar(y_pos, bin_count, align='center', color='orange', alpha=0.7)
plt.xticks(y_pos, labels)
plt.xlabel("Query length", fontsize=17, labelpad=20)
plt.ylabel("Number of sequences", fontsize=1, labelpad=20)
plt.savefig("query_len.png")
plt.show()
```
